```python
import cobra
from ssbio.pipeline.gempro import GEMPRO
import sys
import logging
from os.path import join
import pandas as pd
import requests
```

```python
logger = logging.getLogger()
logger.setLevel(logging.INFO)
handler = logging.StreamHandler(sys.stderr)
formatter = logging.Formatter('[%(asctime)s] [%(name)s] %(levelname)s: %(message)s', datefmt="%Y-%m-%d %H:%M")
handler.setFormatter(formatter)
logger.handlers = [handler]
```

```python
model_dir='/Applications/labfile/senior_design/models'
data_dir='/Applications/labfile/senior_design/data'
```

```python
model=cobra.io.load_json_model(join(model_dir,"iAF1260.json"))
```

```python
protein_in_cyto_df=pd.read_csv(join(data_dir,'all_proteins__which_localize_to_the_compartment_cytosol.csv'))
gene_list=list(protein_in_cyto_df.Accession_1)
LIST_OF_GENES=[]
for i in range(len(gene_list)):
    if gene_list[i] not in LIST_OF_GENES:
        LIST_OF_GENES.append(gene_list[i])
```

```python
ecoli_cyto = GEMPRO(gem_name='Ecoli cytoplasm', root_dir=data_dir, genes_list=LIST_OF_GENES, pdb_file_type='pdb')
```

```python
ecoli_cyto.uniprot_mapping_and_metadata(model_gene_source='ENSEMBLGENOME_ID')
print('Missing UniProt mapping: ', ecoli_cyto.missing_uniprot_mapping)
ecoli_cyto.df_uniprot_metadata.head()
```

```python
ecoli_cyto.set_representative_sequence()
print('Missing a representative sequence: ', ecoli_cyto.missing_representative_sequence)
ecoli_cyto.df_representative_sequences.head()
```
